#### Univariate ,Bivariate and Mutivariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
df.head(70)
```

```python
df.shape
```

```python
df.tail()
```

### univariate Analysis

```python
df_setosa=df.loc[df['Species']=='Iris-setosa']
```

```python
df_virginica=df.loc[df['Species']=='Iris-virginica']
df_versicolor=df.loc[df['Species']=='Iris-versicolor']
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'o')
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'o')
plt.xlabel('sepallength')
plt.show()
```

```python
### Bivariate Analysis
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend();
plt.show()
```

```python
### Bivariate Analysis
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,'PetalLengthCm','SepalWidthCm').add_legend();
plt.show()
```

```python
### Bivariate Analysis
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,'PetalLengthCm','PetalLengthCm').add_legend();
plt.show()
```

```python
### Multivariate Analysis
sns.pairplot(df,hue='Species',size=3)
plt.show()
```
